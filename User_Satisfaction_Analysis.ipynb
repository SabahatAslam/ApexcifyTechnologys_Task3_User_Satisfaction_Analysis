### 🔑Key Concepts

- Data visualization, groupby().size(), value_counts(), unstack(fill_value=0), stacked barplot(), xlabel(), ylabel(), title(), legend()

```python
import pandas as pd
customer_satisfaction_dataset = pd.read_csv("/content/user_satisfaction_150.csv")
customer_satisfaction_dataset
```

🔍 Let’s count the satisfaction levels selected by customers.

>📌The **value_counts()** function counts how many times each value appears in a column.

💡 It is especially useful for analyzing survey data and understanding response distribution.


```python
satisfied_user_count = customer_satisfaction_dataset["satisfaction"].value_counts()
satisfied_user_count
```

## 📈 Visualizing Satisfaction Data:
📊 Now that we understand the value counts, we can easily visualize this data using powerful data visualization libraries.

🚀 Commonly used libraries include:

- 📉 Matplotlib  
- 🎨 Seaborn  

We will plot using seaborn and customize/label it with matplotlib.


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x =satisfied_user_count.index,  y= satisfied_user_count.values) # .index gives us the names/labels & .values gives us the actual values/data
plt.title("Customer Satisfaction Distribution")
plt.xlabel("Satisfaction Level")
plt.ylabel("Number of users")
plt.xticks(rotation = 30) # To avoid smudging of x-axis data labels
plt.show()
```

### 🛒Customer Satisfaction by Age Group

Let’s analyze the customer satisfaction by age group and gain insights into customer behavior.

>💡This can help target marketing strategies with respect to age segments.

```python
customer_by_age =customer_satisfaction_dataset["age"].value_counts()
customer_by_age
```

### 👥Grouping Customers by Age & Satisfaction
We can group the customers of same age based on satisfaction.
- .groupby() will put same age group with same satisfaction level into one group
- .size() will count how many of same age group have same satisfaction level 🔢.

>💡 This will helps us see exactly how many people of each age have same satisfaction level 📊

```python
age_satisfaction = customer_satisfaction_dataset.groupby(["age", "satisfaction"]).size()
age_satisfaction
```

### 📊Making the Table Readable for Visualization

We can reshape our grouped data for better readability and plotting:

- .unstack() converts a row index into column index, making the table easier to read 🗂️
- fill_value=0 replaces missing or NaN values with 0.




```python
age_satisfaction = customer_satisfaction_dataset.groupby(["age", "satisfaction"]).size().unstack(fill_value=0)
age_satisfaction
```

### 🔍Bivariate Analysis:
>💡 Since the data is multidimensional (age × satisfaction x count), a stacked bar chart is the ideal way to visualize it 📈

where, we will have age on x axis, number of customers on y-axis and satisfaction levels stacked along y-axis.

```python
age_satisfaction.plot(kind="bar", stacked=True, figsize=(15,8)) # Bar for vertical bars & barh for horizontal bars.
# we want stacked graph, we write true. If we write false, bars are placed side by side.
plt.title("Age vs Satisfaction Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.legend(title="Satisfaction")


plt.show()
```
